```python
import pandas as pd 

economic_df = pd.read_csv("econ_data.csv")

#We need to convert years into election cycle to match datasets with no NAs
economic_df['Year'] = ((economic_df['Year']- 1948) // 4) * 4 + 1948
election_df = pd.read_csv("election_data.csv")
#rename the column year to prepare for a left join on year
election_df = election_df.rename(columns={'year': 'Year'})
```

```python
#We use the merge function to combine datasets on year
merged_df = economic_df.merge(election_df, on = 'Year', how = 'left')
merged_df.head(40)
```

```python
merged_df = merged_df[['State', 'Total Votes', 'Republican Votes', 'Republican Percentage', 'Republican Electoral',
                       'Democratic Votes', 'Democratic Percentage', 'Democratic Electoral', 'President Elect', 
                       'Winning Party', 'UNRATE', 'GDP', 'CPI', 'Month', 'Year']]
```

```python
merged_df = merged_df.rename(columns={'Year': 'Election Cycle'})
merged_df.head()
```

```python
merged_df.to_csv("economic_election_data.csv")
```
